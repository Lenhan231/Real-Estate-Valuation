# 📊 Phân Tích Dữ Liệu Khám Phá (EDA)
## Real Estate Valuation Dataset

Phân tích toàn diện dữ liệu bất động sản từ Supabase

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
import sys

sys.path.insert(0, str(Path('.').resolve().parent))
from pipeline.supabase_handler import fetch_csv_from_supabase

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline

print('✅ Imports thành công!')
```

## 1. Load Dữ Liệu từ Supabase

```python
print('Loading dữ liệu từ Supabase...')
df = fetch_csv_from_supabase()
print(f'✅ Đã load {len(df)} records')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Tổng Quan Dữ Liệu

```python
print('Data Types:')
print(df.dtypes)
print('Missing Values:')
print(df.isnull().sum())
```

## 3. Phân Tích Giá (Price Analysis)

```python
print('Thống kê GIÁ:')
print(df['price'].describe())
print(f'Khoảng giá: {df["price"].min():,.0f} - {df["price"].max():,.0f} VND')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].hist(df['price'], bins=50, edgecolor='black', color='skyblue')
axes[0, 0].set_title('Phân Bố Giá')
axes[0, 0].set_xlabel('Giá (VND)')

axes[0, 1].hist(np.log10(df['price']), bins=50, edgecolor='black', color='orange')
axes[0, 1].set_title('Phân Bố Giá (Log Scale)')
axes[0, 1].set_xlabel('Log10(Giá)')

axes[1, 0].boxplot(df['price'], vert=True)
axes[1, 0].set_title('Box Plot')
axes[1, 0].set_ylabel('Giá (VND)')

sorted_price = np.sort(df['price'])
cdf = np.arange(1, len(sorted_price) + 1) / len(sorted_price)
axes[1, 1].plot(sorted_price, cdf, linewidth=2, color='green')
axes[1, 1].set_title('Hàm Phân Bố Tích Lũy')
axes[1, 1].set_xlabel('Giá (VND)')

plt.tight_layout()
plt.savefig('price_distribution.png', dpi=100, bbox_inches='tight')
plt.show()
print('✅ Đã lưu: price_distribution.png')
```

## 4. Tương Quan Đặc Trưng

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print('Tương Quan với GIÁ:')
correlations = df[numeric_cols].corr()['price'].sort_values(ascending=False)
print(correlations)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
top_features = correlations.abs().nlargest(15).index.tolist()
sns.heatmap(df[top_features].corr(), annot=True, fmt='.2f', cmap='coolwarm',
            center=0, ax=ax, cbar_kws={'label': 'Tương Quan'})
ax.set_title('Ma Trận Tương Quan (Top 15 Đặc Trưng)')
plt.tight_layout()
plt.savefig('correlation_heatmap.png', dpi=100, bbox_inches='tight')
plt.show()
print('✅ Đã lưu: correlation_heatmap.png')
```

## 5. Phân Tích Địa Lý

```python
fig, ax = plt.subplots(figsize=(12, 10))
scatter = ax.scatter(df['lon'], df['lat'], c=df['price'], cmap='viridis',
                     s=30, alpha=0.6)
ax.set_xlabel('Kinh Độ')
ax.set_ylabel('Vĩ Độ')
ax.set_title('Bản Đồ Nhiệt Giá')
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Giá (VND)')
plt.tight_layout()
plt.savefig('price_heatmap.png', dpi=100, bbox_inches='tight')
plt.show()
print('✅ Đã lưu: price_heatmap.png')
```
